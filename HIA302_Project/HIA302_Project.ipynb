# Descriptive Analysis

```python
ae_by_dose = drop_date_combined_df.groupby('vaxtype')[['daily_nonserious_mysj_dose1','daily_nonserious_mysj_dose2']].sum()
ae_by_dose
```

```python
#Total dose 1
Total_D1_admin = [df9_sum.loc['astrazeneca', 'astra1'], df9_sum.loc['pfizer', 'pfizer1'], df9_sum.loc['sinovac', 'sinovac1']]
df9_sum['Total_D1_admin'] = Total_D1_admin

#Total dose 2
Total_D2_admin = [df9_sum.loc['astrazeneca', 'astra2'], df9_sum.loc['pfizer', 'pfizer2'], df9_sum.loc['sinovac', 'sinovac2']]
df9_sum['Total_D2_admin'] = Total_D2_admin

#Total dose for each vax
Total_dose_admin = df9_sum[['Total_D1_admin', 'Total_D2_admin']].sum(axis=1)
df9_sum['Total_dose_admin'] = Total_dose_admin

#drop columns
df9_sum = df9_sum.drop(['pfizer1', 'pfizer2', 'sinovac1','sinovac2', 'astra1', 'astra2'], axis=1)

df9_sum.insert(0, column = "vaxtype", value = df9_sum.index)
df9_sum.to_csv('Summarized_Covid_Vax_SEs.csv')
```

#### Percentange of SE reported for each vaccine respectively. (LiXen)

```python
## (total SE for each vax/ total dose administered for each vax respectively)
TotalSE_TotalDose = df9_sum[['daily_nonserious_mysj_dose1',
       'daily_nonserious_mysj_dose2']].apply(lambda x: percent(x, df9_sum['Total_dose_admin']))
TotalSE_TotalDose.insert(0, column = "vaxtype", value = TotalSE_TotalDose.index)
TotalSE_TotalDose.to_csv('TotalSE_percentage_byVax.csv')

## bargraph: dose 1 and dose 2 % over Total administered, grouped by vaxtype
TotalSE_TotalDose.plot(kind='bar', stacked=True, color=['powderblue', 'palevioletred'], width=0.5)
plt.xlabel("Type of Vaccine")
plt.ylabel("Percentage (%)")
plt.title("Percentage of non-serious side effects \nreported based on Vaccination type", fontsize=16)
plt.legend(labels = ['Dose 1', 'Dose 2'])
```

# Percentage of non-serious side effects reported for dose1 and dose 2 (Li Xen)

```python
## (total SE1/ total D1 ; total SE2/ total D2)
TotalSEoverDose = pd.DataFrame()

TotalSEoverDose['TotalSE1_D1_pct'] = percent(df9_sum['daily_nonserious_mysj_dose1'], df9_sum['Total_D1_admin'])
TotalSEoverDose['TotalSE2_D2_pct'] = percent(df9_sum['daily_nonserious_mysj_dose2'], df9_sum['Total_D2_admin'])
TotalSEoverDose= TotalSEoverDose.transpose()

Data1 = TotalSEoverDose.rename({'TotalSE1_D1_pct':'Dose 1', 'TotalSE2_D2_pct':'Dose 2'})
Data1.plot(kind='bar', stacked=False, color=['skyblue', 'pink', 'plum'])
plt.xlabel("")
plt.ylabel("Percentage (%)")
plt.title("Percentage of non-serious side effects reported \nfor Dose 1 and Dose 2 respectively", fontsize=16)
plt.legend(title='Vaccination Type')
```
