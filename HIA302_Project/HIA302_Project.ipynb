# Descriptive Analysis

```python
#Monthly trend of each vaccine
combined_filtered_df['month'] = pd.DatetimeIndex(combined_filtered_df['date']).month
monthly_totals = combined_filtered_df.groupby([combined_filtered_df['month'], 'vaxtype']).sum().unstack()
plt.figure(figsize=(8,6))
plt.plot(monthly_totals.index, monthly_totals['daily_total']['astrazeneca'],
         label='Astrazeneca', marker='o')
plt.plot(monthly_totals.index, monthly_totals['daily_total']['pfizer'],
         label='Pfizer', marker='o')
plt.plot(monthly_totals.index, monthly_totals['daily_total']['sinovac'],
         label='Sinovac', marker='o')
plt.title('Monthly Trend of Adverse Effects Reported for Each Vaccine Type')
plt.xlabel('Month')
plt.ylabel('Monthly Total Adverse Effects')
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Dataframe to calculate percentage (Li Xen)

```python


# Data manipulation - organising data merged from vax_malaysia.csv and forming new columns
#Total dose 1 administered for each vax
Total_D1_admin = [df9_sum.loc['astrazeneca', 'astra1'], df9_sum.loc['pfizer', 'pfizer1'], df9_sum.loc['sinovac', 'sinovac1']]
df9_sum['Total_D1_admin'] = Total_D1_admin

#Total dose 2 administered for each vax
Total_D2_admin = [df9_sum.loc['astrazeneca', 'astra2'], df9_sum.loc['pfizer', 'pfizer2'], df9_sum.loc['sinovac', 'sinovac2']]
df9_sum['Total_D2_admin'] = Total_D2_admin

#Total of dose 1 and dose 2 of each vax
Total_dose_admin = df9_sum[['Total_D1_admin', 'Total_D2_admin']].sum(axis=1)
df9_sum['Total_dose_admin'] = Total_dose_admin

#drop columns that are no more needed
df9_sum = df9_sum.drop(['pfizer1', 'pfizer2', 'sinovac1','sinovac2', 'astra1', 'astra2'], axis=1)

df9_sum.insert(0, column = "vaxtype", value = df9_sum.index)
df9_sum.to_csv('Summarized_Covid_Vax_SEs.csv')
```

# Percentange of total non-serious AE reported for each vaccine respectively. (LiXen)

```python
## (total SE for each vax/ total dose administered for each vax respectively)
TotalSE_TotalDose = df9_sum[['daily_nonserious_mysj_dose1',
       'daily_nonserious_mysj_dose2']].apply(lambda x: percent(x, df9_sum['Total_dose_admin']))
TotalSE_TotalDose.insert(0, column = "vaxtype", value = TotalSE_TotalDose.index)
TotalSE_TotalDose.to_csv('TotalSE_percentage_byVax.csv')

## bargraph: dose 1 and dose 2 % over Total administered, grouped by vaxtype
ax = TotalSE_TotalDose.plot(kind='bar', stacked=True, width=0.5)
plt.xlabel("Type of Vaccine")
plt.ylabel("Percentage (%)")
plt.title("Percentage of total non-serious AEs reported\nfor each vaccine type", fontsize=16)
plt.legend(labels = ['Dose 1', 'Dose 2'])

for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.1f}%', (x + width/2, y + height/2), ha='center', va='center')

```

# Heatmap - ALL AE vs Vaxtype

```python
vaccine_doses = ['pfizer1', 'astra1','sinovac1']

# Columns for reported side effects (assuming they start with 'd1_' or 'd2_')
symptoms_columns = [col for col in drop_date_combined_df.columns if col.startswith('d1_')]

# Selecting the relevant columns for vaccine doses and side effects
selected_columns = vaccine_doses + symptoms_columns
selected_data = drop_date_combined_df[selected_columns]

# Calculating the correlation matrix for the selected columns
selected_correlation_matrix = selected_data.corr()
```

```python
# Transposing the correlation matrix to have vaccine doses on the y-axis and side effects on the x-axis
transposed_correlation_matrix = selected_correlation_matrix.T[vaccine_doses]

# Creating the adjusted correlation heat map
plt.figure(figsize=(15, 20))
sns.heatmap(transposed_correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm")
plt.title('Correlation Heat Map: Vaccine Doses and Reported Side Effects')
plt.xlabel('Reported Side Effects')
plt.ylabel('Vaccine Doses')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.show()
```
